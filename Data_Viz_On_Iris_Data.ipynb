# TO VISUALIZE THE IRIS DATASET

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
iris = pd.read_csv("E:\Jupyter Notebook\Dataset/Iris.csv")
iris.head()
```

```python
# Renaming the Columns
iris.rename(columns = {'SepalLengthCm': "seplen", 'SepalWidthCm': "sepwid", 'PetalLengthCm': "petlen", 'PetalWidthCm': "petwid", "Species": "specie"}, inplace = True)
```

```python
# Dropping the Id Column to avoid Redundancy
iris.drop("Id", inplace = True, axis=1)
```

```python
iris.head()
```

```python
sns.scatterplot(x="petwid", y="petlen", data=iris)
```

```python
sns.stripplot(x="specie", y="petlen", data=iris)
```

```python
sns.swarmplot(x="specie", y="petlen", data=iris)
```

```python
plt.figure(figsize=(20,8))
sns.countplot(x="petlen", data=iris, palette = "Greens_d")
plt.show()
```

```python
plt.figure(figsize=(20,8))
sns.pointplot(x="petlen", y="seplen", hue='specie', data=iris, 
              markers=["^","o","*"],
              linestyles=["-","--","-."])
```

```python
sns.boxplot(x='petwid',y="sepwid", data=iris)
```

```python
plt.figure(figsize=(20,10))
sns.boxplot(data=iris, orient="h")
```

```python
sns.barplot(x="sepwid", y="petlen", hue="specie", data=iris)
```

```python
sns.heatmap(iris.corr(), annot=True)
```
